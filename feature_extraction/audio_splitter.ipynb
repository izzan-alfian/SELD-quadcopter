```python
import librosa
import numpy as np
import pandas
import soundfile as sf
from IPython.display import Audio
from scipy.io.wavfile import write
```

```python
sound, fs = librosa.load('./resources/[_0SMVzXhf-s]_cut.wav', sr=None)
```

```python
hop_lenght = 512
window_size = 1024
decible_treshold = -30.0

whistle_sounds_indices = []

i = 0
while i < (len(sound) - window_size + 1):
    frame = sound[i : i + window_size]
    rms = np.sqrt(np.mean(frame ** 2))
    decible = librosa.amplitude_to_db(rms)

    whistle_sound = []
    while decible > decible_treshold:
        whistle_sound.extend(np.arange(i, i + hop_lenght, 1))
        i += hop_lenght

        frame = sound[i : i + window_size]
        rms = np.sqrt(np.mean(frame ** 2))
        decible = librosa.amplitude_to_db(rms)

    if len(whistle_sound) > (fs / 2): # if the sound is longer than half a second
        whistle_sounds_indices.append(whistle_sound)
    else:
        i += hop_lenght
```

```python
len(whistle_sounds_indices)
```

```python
Audio(sound[whistle_sounds_indices[0]], rate=fs)
```

```python
for index, i in enumerate(whistle_sounds_indices):
    sf.write(('./whistles/[_0SMVzXhf-s]_whistle_sound_%d.wav' %(index + 1)), sound[i], fs)
```
