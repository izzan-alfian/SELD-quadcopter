```python
import numpy as np
import pyroomacoustics as pra
import matplotlib.pyplot as plt
from scipy.io import wavfile
from IPython.display import Audio
```

```python
room = pra.AnechoicRoom(fs=16000, air_absorption=True)
```

```python
source_position = [5, 0, 5]  # Position of sound source (in meters)
fs, signal = wavfile.read("arctic_a0010.wav")
room.add_source(source_position, signal=signal)
```

```python
mic_positions = np.array([
    [ 0.0420,  0.0615, -0.0410],
    [-0.0420,  0.0615,  0.0410],
    [-0.0615,  0.0420, -0.0410],
    [-0.0615, -0.0420,  0.0410],
    [-0.0420, -0.0615, -0.0410],
    [ 0.0420, -0.0615,  0.0410],
    [ 0.0615, -0.0420, -0.0410],
    [ 0.0615,  0.0420,  0.0410],
    ]).T  # Position of microphone


room.add_microphone_array(mic_positions)
```

```python
room.compute_rir()
room.simulate()
```

```python
n_mic, n_samp = room.mic_array.signals.shape
white = np.random.randn(n_mic, n_samp)
sound = room.mic_array.signals + white * 5
```

```python
Audio(sound, rate=room.fs)
```
